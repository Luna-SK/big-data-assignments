```python
import requests
from lxml import etree
```

```python
url = "https://coding.imooc.com/?sort=0&unlearn=0&page={}"
```

```python
courses = []
for i in range(1, 11):
    url = url.format(i)
    res = requests.get(url)
    text = res.text
    tree = etree.HTML(text)
    courses.extend(tree.xpath(
        "//p[starts-with(@class, 'title')]/text()"))
```

```python
for course_name in courses:
    print(course_name)
```
