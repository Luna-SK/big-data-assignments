```python
names = []
prices = []
types = []
areas = []
for i in range(1, 7):
    urli = url.format(i)
    resp = requests.get(urli, headers=headers)
    tree = etree.HTML(resp.text)
    divs = tree.xpath('//div[@class="resblock-desc-wrapper"]')
    for div in divs:
        names.append(div.xpath('./div[@class="resblock-name"]/a/@title')[0])
        prices.append(div.xpath('.//div[@class="main-price"]/span[@class="number"]/text()')[0])
        types.append(div.xpath('.//a[@class="resblock-room"]/*/text()'))
        areas.append(div.xpath('.//a[@class="resblock-location"]/text()')[1].strip().split('/')[0])
```

```python
cc = []
for a in types:
    if a:
        cc.append(a[-1].split()[1][:-1])
    else:
        cc.append('')
```

```python
dd = []
for a in types:
    if a:
        dd.append(''.join(a[1: -1]))
    else:
        dd.append('0')
```

```python
df = pd.DataFrame({'name': names, 'price': prices, 'area': areas, 'cc': cc, 'dd': dd})
```

```python
df.to_excel('hz.xlsx')
```
